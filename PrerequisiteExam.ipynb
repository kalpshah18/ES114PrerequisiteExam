<h2>Question 1: List Operations and Basic Statistics</h2>
<p><ul>
    <li>Create a list of the first 20 natural numbers.</li>
    <li>Compute the sum of all even numbers in the list.</li>
    <li>Compute the product of all odd numbers in the list.</li>
    <li>Calculate the mean, median, and standard deviation of the list.</li>
</ul></p>

```python
# Part 1
arr = list(range(1, 21))
print("List of First 20 Natural Numbers:", arr)

# Part 2 and 3
evenSum = 0
oddProduct = 1
for num in arr:
    if num % 2 == 0:
        evenSum += num
    else:
        oddProduct *= num
print("Sum of Even Numbers:", evenSum)
print("Product of Odd Numbers:", oddProduct)

# Part 4
n = len(arr)
s = sum(arr)
mean = s / n
print("Mean:", mean)
if n % 2 == 1:
    print("Median:", arr[n // 2])
else:
    n1 = arr[n // 2 - 1] # Indices start from 0 so we had to subtract 1
    n2 = arr[n // 2]
    print("Median:", (n1 + n2) / 2)
deviationSum = 0
for num in arr:
    deviationSum += abs(num - mean)
stdDeviation = deviationSum / n
print("Standard Deviation:", stdDeviation)
```

<h2>Question 3: Fibonacci Sequence</h2>
<p>We will use Matplotlib Library for Plotting the Graph</p>

```python
def fibonacciList(n: int) -> list:
    ls = [0, 1]
    if n == 1:
        return [0]
    for i in range(n - 2):
        ls.append(ls[-1] + ls[-2])
    return ls
```

<h2>Question 9: Merging Dictionaries</h2>
<p>We cannot use '|' Operator Because '|' Operator overwrites Value for Same Key with Value of Second Dictionary. Thus, we have to iterate over each item of both dictionaries.</p>

```python
def mergeDictionaries(dict1: dict, dict2: dict) -> dict:
    result = {}
    for i in dict1.items():
        result[i[0]] = result.get(i[0], 0) + i[1]
    for i in dict2.items():
        result[i[0]] = result.get(i[0], 0) + i[1]
    return result

# Example
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
print("Dictionary 1:", dict1)
print("Dictionary 2:", dict2)
print("Merged Dictionary", mergeDictionaries(dict1, dict2))
```

<h2>Question 11: Remove Duplicates</h2>
<p>As of Python 3.7, Dictionaries have become ordered. We will use this Property to Answer the Question</p>

```python
def removeDuplicates(l: list) -> list:
    d = {}
    for element in l:
        d[element] = d.get(element, 0) + 1
    return list(d.keys())

def mean(l:list) -> float:
    s = sum(l)
    n = len(l)
    return s / n

def median(l: list):
    l.sort()
    n = len(l)
    if n % 2 == 1:
        return l[n // 2] # Indices start from 0
    else:
        n1 = l[n // 2 - 1] # Indices start from 0 so we had to subtract 1
        n2 = l[n // 2]
        return (n1 + n2) / 2
```

```python
# Example Test Case
ls = [1, 2, 3, 3, 4, 4, 5]
print("Original List:", ls)
ls = removeDuplicates(ls)
print("Duplicates Removed from List:", ls)
print("Mean of New List:", mean(ls))
print("Median of New List:", median(ls))
```

<h2>Question 12: String Operations</h2>

```python
def strOperations(s: [str]):
    longestStr = ''
    longestStrLen = 0
    vowelStartCount = 0
    stringLengths = []
    for string in s:
        stringLen = len(string)
        if string[0] in 'aeiou':
            vowelStartCount += 1
        if stringLen > longestStrLen:
            longestStrLen = stringLen
            longestStr = string
        stringLengths.append(stringLen)
    print("Longest String:", longestStr)
    print("Number of Strings starting with a Vowel:", vowelStartCount)

    import matplotlib.pyplot as plt

    plt.title("String Lengths' Frequency")
    plt.xlabel("Length of String")
    plt.ylabel("Frequency")
    plt.hist(stringLengths)
    plt.show()
    plt.close()
```

<h2>Question 15: Anagram Finder</h2>
<p>We will Sort the Letters Alphabetically then for all Words, check the Sorted Version of those Words and if they match then they are Anagrams.</p>

```python
def anagramFinder(word: str, ls: [str]) -> [str]:
    output = []
    sortedWord = sorted(word)
    for w in ls:
        if sorted(w) == sortedWord:
            output.append(w)
    return output
```

```python
# Example Test Case
word = "listen"
words = ["enlist", "google", "inlets", "banana"]
print(anagramFinder(word, words))
```

<h2>Question 16: Set Operations</h2>

```python
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

def setOperations(s1, s2):
    intersection = s1.intersection(s2)
    union = s1.union(s2)
    symmetric_difference = s1.symmetric_difference(s2)
    return intersection, union, symmetric_difference

def visualizeResults(ls1, ls2):
    s1 = set(ls1)
    s2 = set(ls2)
    
    results = list(setOperations(s1, s2))
    print("Intersection:", results[0])
    print("Union:", results[1])
    print("Symmetric Difference:", results[2])
    
    venn2([s1, s2], set_labels=('List 1', 'List 2'))
    plt.title('Set Operations Visualization')
    plt.show()
    plt.close()
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

visualizeResults(list1, list2)
```

<h2>Question 17: File Operations</h2>

```python
def fileOperations(ls: [int]) -> int:
    with open('data.txt', 'wt') as fs:
        for num in ls:
            fs.write(str(num) + '\n')
    fs.close()
    with open('data.txt', 'rt') as fsRead:
        s = 0
        while True:
            try:
                s += int(fsRead.readline())
            except:
                break
        return s
```

```python
# Example Test Case
print(fileOperations([1,2,3,4,5]))
```

<h2>Question 18: Word Count and Frequency</h2>

```python
import matplotlib.pyplot as plt

def wordCount():
    with open('file.txt', 'rt') as fs:
        data = fs.read()
        words = data.split()
        frequency = {}
        for word in words:
            frequency[word] = frequency.get(word, 0) + 1
        maxFreq = 0
        maxFreqWord = 0
        for key, value in frequency.items():
            if value > maxFreq:
                maxFreq = value
                maxFreqWord = key
        print("Word with Maximum Frequency is:", maxFreqWord)

        plt.title("Word Frequency")
        plt.xlabel("Words")
        plt.ylabel("Frequency")
        plt.bar(list(frequency.keys())[:10], list(frequency.values())[:10]) # Showing only Top 10 Frequencies or else Graph will get too crowded
        plt.show()
        plt.close()

wordCount()
```

<h2>Question 21: Basic Statistics</h2>

```python
def statistics(ls: [int]):
    s = sum(ls)
    n = len(ls)
    mean = s / n

    median = 0
    ls.sort()
    if n % 2 == 1:
        median = ls[n // 2] # Indices start from 0
    else:
        n1 = l[n // 2 - 1] # Indices start from 0 so we had to subtract 1
        n2 = l[n // 2]
        median = (n1 + n2) / 2

    frequency = {}
    for num in ls:
        frequency[num] = frequency.get(num, 0) + 1
    maxVal = 0
    mode = 0
    for key, value in frequency.items():
        if value > maxVal:
            maxVal = value
            mode = key

    stdDevSum = 0
    for num in ls:
        stdDevSum += abs(num - mean)
    stdDeviation = stdDevSum / n
    return mean, median, mode, stdDeviation
```

<h2>Question 22: Normalization</h2>

```python
def normalize(ls: [int]) -> [int]:
    s = sum(ls)
    normalizedList = [i/s for i in ls]
    return normalizedList
```

```python
import matplotlib.pyplot as plt

l = [99, 23, 40, 76, 57, 13, 44, 11, 37, 6, 83, 41, 1, 37, 63, 23, 14, 26, 31, 21, 58, 20, 47, 40, 90, 83, 38, 66, 49, 13, 41, 51, 35, 39, 20, 90, 39, 67, 32, 10, 83, 75, 84, 66, 7, 28, 9, 67, 85, 97]
normalizedList = normalize(l)

plt.title("Data Line Plot")
plt.xlabel("Number")
plt.ylabel("Normalized Value")
plt.plot(l, normalizedList)
plt.show()
plt.close()
```
